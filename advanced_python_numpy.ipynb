## Zeroes, ones and eyes

```python
zero = np.zeros(10)  # array of 10 zeros
zero
```

```python
np.zeros((10,3))  # array of zeroes with 10 rows, 3 columns
```

```python
np.ones((3,3)) # array of ones with shape3X3
```

```python
np.eye(5,3) # Return a 2-D array with ones on the diagonal and zeros elsewhere.
```

```python
np.identity(5)
```

**np.identity returns a square matrix** (special case of a 2D-array) which is an identity matrix with the main diagonal (i.e. 'k=0') as 1's and the other values as 0's. you can't change the diagonal k here.

**np.eye returns a 2D-array**, which fills the diagonal, i.e. 'k' which can be set, with 1's and rest with 0's.

So, the main advantage depends on the requirement. If you want an identity matrix, you can go for identity right away, or can call the np.eye leaving the rest to defaults.

But, if you need a 1's and 0's matrix of a particular shape/size or have a control over the diagonal you can go for eye method.

Just like how a matrix is a special case of an array, np.identity is a special case of np.eye.

## Reshape and resize functions

reshape() does not change your data, but resize() does change it. resize() first accommodates all the values in the original array. After that, if extra space is there (or size of new array is greater than original array), it adds its own values.

```python
nd_arr = np.arange(50)
nd_arr
```

```python
len(nd_arr)
```

```python
nd_arr.reshape(5,10)
```

```python
nd_arr.shape # shape is still 50,1
```

```python
nd_arr.resize(5,10)
```

```python
nd_arr.shape  # Here, shape changed to 5X10
```

## Generating random values

numpy.random.randn generates samples from the normal distribution, while numpy.random.rand from a uniform distribution (in the range [0,1)).

#### Using **random.rand** - from a uniform distribution (in the range [0,1))

```python
np.random.rand(5, 3)
```

#### Using **random.randn**   - from a normal distribution

```python
np.random.randn(5, 3)
```

#### Using **random.randint**  Returns a random integer from low,high interval given

```python
np.random.randint(5, 40)
```

```python
np.random.randint(0, 40, 10) # 10 random numbers between 0 and 40
```

```python
np.random.randint(0, 40, (2,5))
```

## Deep and Shallow copy

In Python, we use = operator to create a copy of an object. You may think that this creates a new object; it doesn't. It only creates a new variable that shares the reference of the original object.

```python
array_new = array1
print(id(array_new), id(array1))
```

The copy owns the data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy

```python
array_new = array1.copy()
print(id(array_new), id(array1))
```
